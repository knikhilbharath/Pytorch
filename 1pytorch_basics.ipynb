```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = torch.tensor([1, 2, 3, 4, 5, 6])
y = torch.tensor([5, 6, 7, 8, 9, 1])
z = torch.dot(x, y)
print(z)
```

```python
x = torch.linspace(1, 10, 100)
y = torch.exp(x)
```

```python
plt.plot(x.numpy(), y.numpy())
```

```python
x = torch.linspace(1, 10, 100)
y = torch.sin(x)
```

```python
plt.plot(x.numpy(), y.numpy())
```
